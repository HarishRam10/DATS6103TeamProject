```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Possible project data/covid_19_data.csv")
df.head()
```

```python
df = df[['ObservationDate', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
df.head()
```

```python
df.rename(columns = {'ObservationDate' : 'Obsdate', 'Country/Region' : 'Country'}, inplace = True)
df.head()
```

```python
countries = pd.unique(df['Country'])
```

```python
x = []
y = []
z = []
```

```python
for c in countries:
    confirmed = sum(df[df['Country'] == c].Confirmed)
    death = sum(df[df['Country'] == c].Deaths)
    recovered = sum(df[df['Country'] == c].Recovered)
    x.append(confirmed)
    y.append(death)
    z.append(recovered)
```

```python
df1 = pd.DataFrame({"Country" : countries, 'Total_confirmed' : x, 'Total_deaths' : y, 'Total_recovered' : z})
```

```python
df1.head()
```

```python
df1 = df1.sort_values(by = 'Total_confirmed', ascending = 0)
```

```python
top_fifteen = df1['Country'].head(15)
```

```python
for x in top_fifteen:
    globals()["%s" % x] = df[df['Country'] == x]
    globals()["%s" % x] = globals()["%s" % x].groupby(globals()["%s" % x]['Obsdate']).aggregate({'Obsdate' : 'first', 'Country' : 'first', 'Confirmed' : 'sum', 'Deaths' : 'sum', 'Recovered' : 'sum'})
    globals()["%s" % x]['Cumulative_Confirmed'] = globals()["%s" % x]['Confirmed'].cumsum(axis = 0)
    globals()["%s" % x]['Cumulative_Deaths'] = globals()["%s" % x]['Deaths'].cumsum(axis = 0)
    globals()["%s" % x]['Cumulative_Recovered'] = globals()["%s" % x]['Recovered'].cumsum(axis = 0)
    globals()["%s" % x]['Fatality_Rate'] = globals()["%s" % x]['Cumulative_Deaths']/globals()["%s" % x]['Cumulative_Confirmed']*100
    globals()["%s" % x]['Recovery_Rate'] = globals()["%s" % x]['Cumulative_Recovered']/globals()["%s" % x]['Cumulative_Confirmed']*100
```
