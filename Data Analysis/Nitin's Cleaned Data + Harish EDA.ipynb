<h1> ANALYSIS OF TREND IN COVID-19 CASES </h1>

```python
import pandas as pd
import numpy as np
```

<h2>Importing Dataset</h2>

```python
confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveries_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
print(confirmed_df.shape)
print(deaths_df.shape)
print(recoveries_df.shape)
```

```python
confirmed_df
```

<h2> Data Preprocessing </h2>

<h3> Combining rows under same country. </h3>

```python
df_c = confirmed_df.groupby('Country/Region').sum()
df_c
```

```python
df_d = deaths_df.groupby('Country/Region').sum()
df_d
```

```python
df_r = recoveries_df.groupby('Country/Region').sum()
df_r
```

<h3> Taking transpose of the datasets for better dataset usage. </h3>

```python
test_c = df_c.transpose()
test1_c = test_c.iloc[2:]
test1_c
```

```python
test_d = df_d.transpose()
test1_d = test_d.iloc[2:]
test1_d
```

```python
test_r = df_r.transpose()
test1_r = test_r.iloc[2:]
test1_r
```

<h3> Creating the main dataset. </h3>

```python
DF = pd.DataFrame(columns = ['Date','Country','Confirmed','Deaths','Recovered','Cumulative_Confirmed','Cumulative_Deaths','Cumulative_Recovered','Fatality_Rate','Recovery_Rate'])
```

```python
for i in range(len(test1_c.columns)):
    for j in range(len(test1_c.index)):
        d = test1_c.index[j]
        c = test1_c.columns[i]
        conf = test1_c[c][j]
        dth = test1_d[c][j]
        rcvd = test1_r[c][j]
        temp = pd.DataFrame({'Date' : d, 'Country' : c, 'Confirmed' : conf, 'Deaths' : dth, 'Recovered' : rcvd}, index = [(i*len(test1_c.index))+j])
        DF = pd.concat([DF,temp])
```

```python
DF
```

```python
DF['Cumulative_Confirmed'] = DF.groupby(['Country','Date'],level=0).cumsum()['Confirmed']
DF['Cumulative_Deaths'] = DF.groupby(['Country','Date'],level=0).cumsum()['Deaths']
DF['Cumulative_Recovered'] = DF.groupby(['Country','Date'],level=0).cumsum()['Recovered']
```

```python
DF.head(311)
```

```python
f = DF.groupby(['Country','Date'],level=0).apply(lambda x : (x['Cumulative_Deaths']/x['Cumulative_Confirmed']*100) if x['Cumulative_Deaths'].item() > 0 else 0)
rr = DF.groupby(['Country','Date'],level=0).apply(lambda x : (x['Cumulative_Recovered']/x['Cumulative_Confirmed']*100) if x['Cumulative_Recovered'].item() > 0 else 0)
```

```python
for i in range(len(DF)):
    if type(f[i]) == pd.core.series.Series:
        DF['Fatality_Rate'].iloc[i] = f[i].iloc[0]
    else:
        DF['Fatality_Rate'].iloc[i] = f[i]
    if type(rr[i]) == pd.core.series.Series:
        DF['Recovery_Rate'].iloc[i] = rr[i].iloc[0]
    else:        
        DF['Recovery_Rate'].iloc[i] = rr[i]
```

```python
map_df = pd.DataFrame(data = DF.groupby('Country')['Cumulative_Confirmed'].max(), columns = ['Cumulative_Confirmed'])
map_df = map_df.sort_values(by=['Cumulative_Confirmed'],ascending = False).head(15)
map_df
```
